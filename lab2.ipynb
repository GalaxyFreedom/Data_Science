```python
# 2.1 Load data set IRIS (https://archive-beta.ics.uci.edu/dataset/53/iris) by using NumPy.
import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_data = np.genfromtxt(url, delimiter=',', usecols=[0,1,2,3])
```

```python
# 2.2 Compute and report eight descriptive metrics (max, min, mean, standard
# deviation, variance, median, .25 quartile, .75 quartile) of IRIS for all the
# flower types.

# Compute the metrics
# max values for each column
max_values = np.amax(iris_data, axis=0)
# min values for each column
min_values = np.amin(iris_data, axis=0)
# mean values for each column
mean_values = np.mean(iris_data, axis=0)
# standard deviation for each column
std_values = np.std(iris_data, axis=0)
# variance for each column
var_vales = np.var(iris_data, axis=0)
# median values for each column
median_values = np.median(iris_data, axis=0)
# 25th percentile for each column
percentile_25 = np.percentile(iris_data, 25, axis=0)
# 75th percentile for each column
percentile_75 = np.percentile(iris_data, 75, axis=0)

# Print the results
print('Max values: ', max_values)
print('Min values: ', min_values)
print('Mean values: ', mean_values)
print('Standard deviation: ', std_values)
print('Variance: ', var_vales)
print('Median values: ', median_values)
print('25th percentile: ', percentile_25)
print('75th percentile: ', percentile_75)
```

```python
# 2.4 Compute and report eight descriptive metrics (max, min, mean, standard
# deviation, variance, median, .25 quartile, .75 quartile) of IRIS for each
# flower type.
import pandas as pd

# Set the URL for the IRIS dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Read the data from the URL into a DataFrame
iris_data = pd.read_csv(url, header=None)

# Create a separate DataFrame for each flower type
# Use boolean indexing to select only the rows where the value in the fifth column is equal to the desired flower type
setosa = iris_data[iris_data[4] == 'Iris-setosa'].iloc[:, :4]
versicolor = iris_data[iris_data[4] == 'Iris-versicolor'].iloc[:, :4]
virginica = iris_data[iris_data[4] == 'Iris-virginica'].iloc[:, :4]

# Compute the metrics for each flower type
for flower_type, data in zip(['Setosa - ', 'Versicolor - ', 'Virginica - '], [setosa, versicolor, virginica]):
    # Print the flower type
    print(flower_type)
    
    # Compute and print the maximum value for each column
    print('Max:', data.max().to_numpy())
    
    # Compute and print the minimum value for each column
    print('Min:', data.min().to_numpy())
    
    # Compute and print the mean value for each column
    print('Mean:', data.mean().to_numpy())
    
    # Compute and print the standard deviation for each column
    print('Standard deviation:', data.std().to_numpy())
    
    # Compute and print the variance for each column
    print('Variance:', data.var().to_numpy())
    
    # Compute and print the median value for each column
    print('Median:', data.median().to_numpy())
    
    # Compute and print the .25 quartile for each column
    print('0.25 quartile:', data.quantile(0.25).to_numpy())
    
    # Compute and print the .75 quartile for each column
    print('0.75 quartile:', data.quantile(0.75).to_numpy())
    
    # Print a blank line to separate the output for each flower type
    print()
```

```python
# 2.5 Draw and report eight descriptive metrics (max, min, mean, standard
# deviation, variance, median, .25 quartile, .75 quartile) of IRIS for each
# flower type.


import matplotlib.pyplot as plt

# Set the URL for the IRIS dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Read the data from the URL into a DataFrame
iris_data = pd.read_csv(url, header=None)

# Create a separate DataFrame for each flower type
# Use boolean indexing to select only the rows where the value in the fifth column is equal to the desired flower type
setosa = iris_data[iris_data[4] == 'Iris-setosa'].iloc[:, :4]
versicolor = iris_data[iris_data[4] == 'Iris-versicolor'].iloc[:, :4]
virginica = iris_data[iris_data[4] == 'Iris-virginica'].iloc[:, :4]

# Compute the metrics for each flower type
for flower_type, data in zip(['Setosa', 'Versicolor', 'Virginica'], [setosa, versicolor, virginica]):
    # Print the flower type
    print(flower_type)
    
    # Compute and print the maximum value for each column
    max_values = data.max().to_numpy()
    print('Max:', max_values)
    
    # Compute and print the minimum value for each column
    min_values = data.min().to_numpy()
    print('Min:', min_values)
    
    # Compute and print the mean value for each column
    mean_values = data.mean().to_numpy()
    print('Mean:', mean_values)
    
    # Compute and print the standard deviation for each column
    std_values = data.std().to_numpy()
    print('Standard deviation:', std_values)
    
    # Compute and print the variance for each column
    var_values = data.var().to_numpy()
    print('Variance:', var_values)
    
    # Compute and print the median value for each column
    median_values = data.median().to_numpy()
    print('Median:', median_values)
    
    # Compute and print the .25 quartile for each column
    q1_values = data.quantile(0.25).to_numpy()
    print('0.25 quartile:', q1_values)
    
    # Compute and print the .75 quartile for each column
    q3_values = data.quantile(0.75).to_numpy()
    print('0.75 quartile:', q3_values)
    
    # Draw a bar chart of the metrics for each column
    x = ['Max', 'Min', 'Mean', 'Std', 'Var', 'Median', 'Q1', 'Q3']
    y = [max_values.mean(), min_values.mean(), mean_values.mean(), std_values.mean(), var_values.mean(), median_values.mean(), q1_values.mean(), q3_values.mean()]
    
    plt.bar(x, y)
    plt.title(flower_type)
    plt.show()
    
    # Print a blank line to separate the output for each flower type
    print()
```

```python
# 2.6 Compute and report the correlation between each pair-wise feature for all
# the flower types

# Set the column names
iris_data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'flower_type']

# Compute and print the correlation matrix
corr_matrix = iris_data.corr()
print(corr_matrix)
```

```python
# 2.7 Compute and report the correlation between each pair-wise feature foR each flower type.

# iris_data['flower_type'].unique() returns an array of the unique values in the ‘flower_type’ column of the iris_data DataFrame. 
# This is used in the for loop to iterate over each unique flower type in the dataset.

# For example, if the ‘flower_type’ column contains the values ‘Iris-setosa’, ‘Iris-versicolor’, and ‘Iris-virginica’, 
# then iris_data['flower_type'].unique() would return an array with these three values: array(['Iris-setosa', 'Iris-versicolor', 
# 'Iris-virginica'], dtype=object). The for loop would then iterate over these three values, setting the value of flower_type to 
# ‘Iris-setosa’ on the first iteration, ‘Iris-versicolor’ on the second iteration, and ‘Iris-virginica’ on the third iteration.
for flower_type in iris_data['flower_type'].unique():
    iris_data_by_type = iris_data[iris_data['flower_type'] == flower_type]
    print(f'Flower type: {flower_type}')
    corr_matrix = iris_data_by_type.corr()
    print(corr_matrix)
    print()
```

```python
# 2.12 Draw and report the distribution chart for each feature of all the flower types
import pandas as pd
import matplotlib.pyplot as plt

iris_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)  # Load iris dataset from URL
iris_data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']  # Assign column names

features = iris_data.columns[:-1]  # Get list of feature column names (excluding the last column which is the species column)

for feature in features:  # Iterate over feature column names
    iris_data.groupby('species')[feature].plot.hist(alpha=0.5)  # Create histogram plot for current feature grouped by species
    plt.title(feature)  # Set plot title to current feature
    plt.legend()  # Show legend
    plt.show()  # Display the plot
```

```python
# 2.18 Draw and report the density chart for each pair-wise feature of all the flower types.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Load the iris dataset from the provided URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = pd.read_csv(url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Create pair-wise density charts for each pair-wise feature of all flower types
features = iris.columns[:-1]
for i in range(len(features)):
    for j in range(i+1, len(features)):
        x = iris[features[i]]
        y = iris[features[j]]
        xy = np.vstack([x,y])
        z = gaussian_kde(xy)(xy)
        iris.plot.scatter(x=features[i], y=features[j], c=z, cmap='viridis')
        plt.title(f'{features[i]} vs {features[j]}')
        plt.show()
```

```python
# 2.19 Draw and report the density chart for each pair-wise feature of each flower type.

# Create density charts for each pair-wise feature of each flower type
features = iris.columns[:-1]
for flower_type in iris['species'].unique():
    iris_by_type = iris[iris['species'] == flower_type]
    for i in range(len(features)):
        for j in range(i+1, len(features)):
            x = iris_by_type[features[i]]
            y = iris_by_type[features[j]]
            xy = np.vstack([x,y])
            z = gaussian_kde(xy)(xy)
            iris_by_type.plot.scatter(x=features[i], y=features[j], c=z, cmap='viridis')
            plt.title(f'{flower_type}: {features[i]} vs {features[j]}')
            plt.show()
```
